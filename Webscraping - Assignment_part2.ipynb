```python
website = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
website
```

```python
matches = []
for b in soup.find_all("td",class_ = "table-body__cell u-center-text"):
    matches.append(b.text)
matches[:20] 
```

```python
match = []
for d in range(0,len(matches),2)[:11]:
    match.append(matches[d])
match
```

```python
rating = []
for e in soup.find_all("td",class_ = "table-body__cell u-text-right rating"):
    rating.append(e.text)
rating[:20] 
```

```python
match = []
for d in range(0,len(matches),2)[:11]:
    match.append(matches[d])
match
```

```python
points = []
for f in range(1,len(matches),2)[:11]:
    points.append(matches[f])
points
```

```python
print(len(teams),len(match),len(points),len(rating))
```

```python
web = requests.get("https://coreyms.com/")
web
```

```python
print(len(heading))
```

```python
date = []
for t in soup3.find_all("time", class_ = "entry-time"):
    date.append(t.text)
date
```

```python
content = []
for s in soup3.find_all("div", class_ = "entry-content"):
    content.append(s.text.replace("\n",""))
content
```

```python
print(len(content))
```

```python
link_list = soup3.find_all('a')
for link in link_list:
    if 'href' in link.attrs:
        print(str(link.attrs['href']))
```
