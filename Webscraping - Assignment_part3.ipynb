```python
from bs4 import BeautifulSoup
import requests
```

# Question : 8

```python
web = requests.get ("https://www.dineout.co.in/chandigarh-restaurants/welcome-back")
web
```

```python
soup = BeautifulSoup(web.content)
soup
```

```python
restaurant = []
for a in soup.find_all("div",class_ = "restnt-info cursor"):
    restaurant.append(a.text)
restaurant
```

```python
cuisine = []
for b in soup.find_all("span",class_ = "double-line-ellipsis"):
    cuisine.append(b.text.split("|")[1])
cuisine
```

```python
location = []
for c in soup.find_all("div",class_ = "restnt-loc ellipsis"):
    location.append(c.text)
location
```

```python
rating = []
for e in soup.find_all("div",class_="restnt-rating rating-4"):
    rating.append(e.text)
rating
```

```python
image = []
for f in soup.find_all("img",class_="no-img"):
    image.append(f["data-src"])
image
```

# Question : 10

```python
website1 = requests.get("https://www.puredestinations.co.uk/")
website1
```

```python
soup2 = BeautifulSoup(website1.content)
soup2
```

```python
monument_name = []
for i in soup2.find_all("div",class_="grid--item__header"):
    monument_name.append(i.text.replace("\n",""))
monument_name [10:]
```

```python
monument_description = []
for j in soup2.find_all("div",class_="grid--item__content"):
    monument_description.append(j.text.replace("\n",""))
monument_description[10:]
```

```python
images = []
for g in soup2.find_all("img",class_="luxury family holidays - luxury holidays - thumbnail"):
    images.append(g["data-src"])
images
```
