```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"C:/Users/Administrator/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
driver.close()
```

```python
url = "https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst&l=bangalore"
```

```python
driver.get(url)
```

```python
job_title=[]
job_location=[]
company_names=[]
experience_required=[]

title=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
title[0:10]

for a in title:
    title=a.text
    job_title.append(title.split('/'))
job_title

tittle = []
for e in range(0,len(job_title),1)[0:10]:
    tittle.append(job_title[e])
tittle

location =driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
location[:10]

for b in location:
    location=b.text
    job_location.append(location)
job_location

locations = []
for f in range(0,len(job_location),1)[0:10]:
    locations.append(job_location[f])
locations

companies_tags= driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags[:10]

for i in companies_tags:
    company_name=i.text
    company_names.append(company_name)
company_names

company = []
for g in range(0,len(company_names),1)[0:10]:
    company.append(company_names[g])
company

experience = driver.find_elements_by_xpath(".//li[@class='fleft grey-text br2 placeHolderLi experience']/span")
experience

for k in experience:
    experiences=k.text
    experience_required.append(experiences)
experience_required

Experience = []
for m in range(0,len(experience_required),1)[0:10]:
    Experience.append(experience_required[m])
Experience

print(len(tittle),len(locations),len(company),len(Experience))

df = pd.DataFrame({"Job_Tittle":tittle,"Location":locations,"Company_name":company,"Experience":Experience})
df
```
